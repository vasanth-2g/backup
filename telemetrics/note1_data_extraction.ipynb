```python
import requests
import zipfile
import os

url = "https://phm-datasets.s3.amazonaws.com/NASA/6.+Turbofan+Engine+Degradation+Simulation+Data+Set.zip"

# Ensure data directory exists
os.makedirs("data", exist_ok=True)

zip_path = "data/data.zip"

# Download the zip file
response = requests.get(url, stream=True)
with open(zip_path, "wb") as f:
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:
            f.write(chunk)

print("ZIP downloaded вњ…")

# Extract contents into data/ folder
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("data")

print("Files extracted into data/ вњ…")
```

```python
import os
import zipfile

for root, dirs, files in os.walk("data"):
    for file in files:
        if file.endswith(".zip"):
            zip_path = os.path.join(root, file)
            print("Extracting:", zip_path)
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall("data")

print("вњ… All extractions done")
```

```python

import pandas as pd

columns = (
    ['unit', 'cycle'] +
    [f'op_setting_{i}' for i in range(1, 4)] +
    [f'sensor_{i}' for i in range(1, 22)]
)
```

```python

train_fd001 = pd.read_csv(
    "data/train_FD001.txt",
    sep=r"\s+",   # handles multiple spaces
    header=None
)

train_fd001.columns = columns

train_fd001.head()
```
